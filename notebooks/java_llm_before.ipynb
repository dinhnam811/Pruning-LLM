````python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "Qwen/Qwen2.5-Coder-3B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    trust_remote_code=True
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

print("Loaded model on:", model.device)
````

````python
import os
import re
import subprocess
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer


# =========================
# 1) Parse signature
# =========================

_sig_re = re.compile(
    r'public\s+static\s+(\S+)\s+([A-Za-z_][A-Za-z0-9_]*)\s*\('
)

def parse_signature(signature: str):
    """
    VÃ­ dá»¥: 'public static boolean isPrime(int n)'
    â†’ return ('boolean', 'isPrime')
    """
    m = _sig_re.search(signature)
    if not m:
        raise ValueError(f"Cannot parse signature: {signature}")
    return_type, name = m.group(1), m.group(2)
    return return_type, name


# =========================
# 2) Gá»i Qwen sinh method Java (Ä‘Ã£ FIX)
# =========================

def generate_java_code(task):
    signature = task["signature"]

    prompt = f"""
You are a Java code generator. STRICT RULES:

1. OUTPUT MUST BE ONLY ONE COMPLETE JAVA METHOD.
2. NO COMMENTS.
3. NO EXPLANATION.
4. NO IMPORT STATEMENTS.
5. ALWAYS CLOSE ALL BRACES.
6. NO EXTRA TEXT BEFORE OR AFTER THE METHOD.
7. NO MARKDOWN. NO ``` BLOCKS.
8. METHOD MUST MATCH SIGNATURE EXACTLY.
9. DO NOT write placeholders like "Your code here", "TODO", empty bodies, or throw exceptions.
10. YOU MUST FULLY IMPLEMENT THE LOGIC SO THAT IT CAN PASS UNIT TESTS.


Signature:
{signature}

Now write ONLY the method:
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        output_ids = model.generate(
            **inputs,
            max_new_tokens=256,
            temperature=0.0,
            do_sample=False,
            pad_token_id=getattr(tokenizer, "eos_token_id", None),
        )

    # ðŸ”´ Chá»‰ láº¥y pháº§n tokens sinh thÃªm (bá» pháº§n prompt)
    gen_ids = output_ids[0][inputs["input_ids"].shape[1]:]

    raw = tokenizer.decode(gen_ids, skip_special_tokens=True).strip()

    # Bá» markdown náº¿u model váº«n cÃ£i lá»i
    raw = raw.replace("```java", "").replace("```", "").strip()

    # Äá» phÃ²ng model echo láº¡i cÃ¢u "Now write ONLY the method:"
    if "Now write ONLY the method:" in raw:
        raw = raw.split("Now write ONLY the method:")[-1].strip()

    # =========================
    # TÃ¬m method Java Ä‘áº§u tiÃªn báº±ng regex
    # =========================
    method_pattern = r"(public\s+static\s+[^\(]+\s+[A-Za-z_][A-Za-z0-9_]*\s*\([^)]*\)\s*\{.*?\})"
    m = re.search(method_pattern, raw, flags=re.DOTALL)

    if m:
        method = m.group(1).strip()
    else:
        # Fallback: tÃ¬m "public static" rá»“i cáº¯t Ä‘áº¿n dáº¥u '}' cuá»‘i cÃ¹ng
        if "public static" not in raw:
            raise ValueError(f"No 'public static' found in model output:\n{raw}")
        snippet = raw[raw.index("public static"):]
        last_brace = snippet.rfind("}")
        if last_brace == -1:
            raise ValueError(f"No closing brace '}}' found in model output:\n{snippet}")
        method = snippet[: last_brace + 1].strip()

    return method


# =========================
# 3) Sinh pháº§n test cho 1 task
# =========================

def generate_test_calls(task):
    signature = task["signature"]
    return_type, method_name = parse_signature(signature)

    calls = []
    for idx, test in enumerate(task["tests"], start=1):
        inp = test["input"]
        expected = test["expected"]

        # primitive types & boolean
        if return_type in ("boolean", "int", "long", "double", "float", "short", "byte", "char"):
            condition = f"{method_name}({inp}) != {expected}"

        # String type
        elif return_type == "String":
            # input / expected pháº£i lÃ  chuá»—i Ä‘Ã£ cÃ³ dáº¥u ngoáº·c kÃ©p trong JSON
            condition = f"!{method_name}({inp}).equals({expected})"

        # fallback (cÃ³ thá»ƒ chá»‰nh sau náº¿u thÃªm kiá»ƒu phá»©c táº¡p)
        else:
            condition = f"{method_name}({inp}) == null"

        call = f"""
        // test #{idx}
        if ({condition}) {{
            throw new Exception("Test {task['id']} failed at case {idx}");
        }}
        """
        calls.append(call)

    return "\n".join(calls)


# =========================
# 4) Build full Main.java cho 1 task
# =========================

def build_java_file(task, method_code):
    tests_code = generate_test_calls(task)

    class_code = f"""
public class Main {{
    {method_code}

    private static void runTests() throws Exception {{
{tests_code}
    }}

    public static void main(String[] args) {{
        try {{
            runTests();
            System.out.println("OK");
        }} catch (Exception e) {{
            System.out.println("FAIL");
        }}
    }}
}}
"""
    return class_code.strip()


# =========================
# 5) Compile + run Java cho 1 task
# =========================

def run_java_test(java_source):
    # luÃ´n ghi Ä‘Ã¨ Main.java trong outputs/code
    filepath = "outputs/code/Main.java"
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(java_source)

    # compile
    comp = subprocess.run(["javac", filepath], capture_output=True, text=True)
    if comp.returncode != 0:
        return False, "compile_error", comp.stderr

    # run
    run = subprocess.run(
        ["java", "-cp", "outputs/code", "Main"],
        capture_output=True,
        text=True
    )
    out = run.stdout.strip()

    if out == "OK":
        return True, "pass", out
    else:
        return False, "fail", out


# =========================
# 6) Eval toÃ n bá»™ tasks â†’ Pass@1
# =========================

def evaluate_pass_at_1(tasks):
    results = []
    num_pass = 0

    print(f"Total tasks: {len(tasks)}")

    for i, task in enumerate(tasks, start=1):
        print(f"\n=== Running task {i}/{len(tasks)}: {task['id']} ===")

        try:
            # 1) Gá»i model sinh method
            method = generate_java_code(task)
            print("\n=== METHOD ===\n", method)

            # 2) Build file Java
            java_src = build_java_file(task, method)
            # Náº¿u cáº§n debug:
            # print("\n=== MAIN.JAVA ===\n", java_src)

            # 3) Compile + run
            passed, status, details = run_java_test(java_src)

        except Exception as e:
            # Báº¥t ká»³ lá»—i nÃ o: extraction, parse, runtime Python
            passed = False
            status = "pipeline_error"
            details = str(e)

        print("\n=== RESULT ===")
        print("Passed:", passed, "| Status:", status)
        print("Details (repr):", repr(details))

        results.append({
            "id": task["id"],
            "passed": passed,
            "status": status,
            "details": details,
        })

        if passed:
            num_pass += 1

    total = len(tasks)
    pass_at_1 = num_pass / total if total > 0 else 0.0

    print("\n=== SUMMARY ===")
    print(f"Pass@1 = {num_pass}/{total} = {pass_at_1:.2%}")

    return pass_at_1, results
````
